```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
combined_features = df.apply(lambda x: ' '.join(x.astype(str)), axis=1)

# Display the combined features
print(combined_features)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer()

# Fit and transform the combined features to create feature vectors
feature_vectors = vectorizer.fit_transform(combined_features)

# Display the feature vectors
print(feature_vectors)
```

```python
# getting the similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
food_name = input(' Enter your favourite food name : ')
```

```python
list_of_all_titles = df['name'].tolist()
print(list_of_all_titles)
```

```python
find_close_match = difflib.get_close_matches(food_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
df['new_id'] = range(1, len(df) + 1)
```

```python
index_of_the_food = df[df.name == close_match]['new_id'].values[0]
print(index_of_the_food)
```

```python
similarity_score = list(enumerate(similarity[index_of_the_food]))
print(similarity_score)
```

```python
sorted_similar_food = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_food)
```

```python
print('Foods suggested for you: \n')

if not df.empty and sorted_similar_food:  # Check if the DataFrame is not empty and sorted_similar_food is not empty
    i = 1
    for food in sorted_similar_food:
        index = food[0]
        if not df[df.new_id == index].empty:  # Check if the filtered DataFrame is not empty
            title_from_index = df[df.new_id == index]['name'].values[0]
            if i < 30:
                print(i, '.', title_from_index)
                i += 1
        else:
            print("No food found.")
            break
else:
    print("No food found.")
```
